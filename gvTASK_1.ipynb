**DATA** **VISUALIZATION**

```python
def add_value_labels(ax,spacing=10):
  for rect in ax.patches:
    x=rect.get_x()+rect.get_width()/2
    y=rect.get_height()-3

    va='bottom' if y>0 else 'top'
    label='()'.format(y)
    y_shift=spacing*(1 if y>=0 else -1)
    ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords='offset points', ha='center',va=va)
```

TOP 10 COUNTRIES OF TOTAL POPULATION IN YEAR 2022 AND 2000

```python
plt.figure(figsize=(20,15))
plt.rcParams['axes.facecolor']=='black'

plt.subplot(2,2,1)
plot=sns.barplot(x="2022",y="Country Name",data=total_highest_population,palette="viridis")
plt.title("Top 10 countries Of Total Population (2022)")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x="2000",y="Country Name",data=total_highest_population, palette="viridis")
plt.title("Top Ten countries of Total Population in 2000")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.tight_layout()
```

TOP 10 COUNTRIES OF TOTAL POPULATION IN YEAR 1999 AND 1989

```python
plt.figure(figsize=(20,15))
plt.rcParams['axes.facecolor']=="black"

plt.subplot(2,2,1)
plot=sns.barplot(x='1999',y="Country Name", data=total_highest_population, palette="mako")
plt.title("Top Ten countries Of Total Population 1999")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x='1989',y="Country Name", data=total_highest_population,palette="mako")
plt.title("Top Ten countries Of Total Population 1989")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.tight_layout()
```

TOP 10 COUNTRIES OF TOTAL POPULATION IN YEAR 1960 AND 2022

```python
plt.figure(figsize=(20,13))
plt.rcParams['axes.facecolor']=="black"

plt.subplot(2,2,1)
plot=sns.lineplot(x='1960',y="Country Name", data=total_highest_population, palette="coolwarm")
plt.title("Top Ten countries Of Total Population 1960")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.lineplot(x='2022',y="Country Name", data=total_highest_population,palette="coolwarm")
plt.title("Top Ten countries Of Total Population 2022")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.tight_layout()
plt.show()
```
